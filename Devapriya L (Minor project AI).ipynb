# With genre

```python
#df.iloc[row_start:row_end, column_start:column_end]
y=df.iloc[:15000,9:10]
y
```

```python
x=df.iloc[:15000,11:17]
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8)
```

```python
xtrain
```

```python
ytrain
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model.fit(xtrain,ytrain)
```

```python
knn_model.predict(xtest)
```

```python
knn_model.score(xtest,ytest)
```

```python
from sklearn.cluster import KMeans
Kmeans_model= KMeans(n_clusters=5)
```

```python
Kmeans_model.fit(df[["energy","playlist_genre"]])
```

```python
Kmeans_model.cluster_centers_
```

```python
df["cluster_group"]=Kmeans_model.labels_
```

```python
df
```

```python
df["cluster_group"].value_counts()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x="energy",y="playlist_genre",data=df,hue="cluster_group")
```
